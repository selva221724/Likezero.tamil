```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
# Display word cloud of titles
from wordcloud import WordCloud

os.chdir('/Users/tamilselvan/Documents/Tamil/Likezero.tamil/data')
# Load the dataset
data = pd.read_csv('True.csv')

# Display basic information about the dataset
print("Basic Information about the Dataset:")
print(data.info())

# Display the first few rows of the dataset
print("\nFirst few rows of the Dataset:")
print(data.head())

# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Explore the distribution of subjects
plt.figure(figsize=(10, 6))
sns.countplot(y='subject', data=data)
plt.title("Distribution of Subjects")
plt.xlabel("Count")
plt.ylabel("Subject")
plt.show()

# Display the distribution of dates
plt.figure(figsize=(10, 6))
data['date'] = pd.to_datetime(data['date'])
data['date'].dt.date.value_counts().sort_index().plot()
plt.title("Distribution of Dates")
plt.xlabel("Date")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()


title_text = ' '.join(data['title'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(title_text)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title("Word Cloud of Titles")
plt.axis('off')
plt.show()
```

```python
# Load the dataset
data = pd.read_csv('Fake.csv')

# Display basic information about the dataset
print("Basic Information about the Dataset:")
print(data.info())

# Display the first few rows of the dataset
print("\nFirst few rows of the Dataset:")
print(data.head())

# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Explore the distribution of subjects
plt.figure(figsize=(10, 6))
sns.countplot(y='subject', data=data)
plt.title("Distribution of Subjects")
plt.xlabel("Count")
plt.ylabel("Subject")
plt.show()

# Display the distribution of dates
plt.figure(figsize=(10, 6))
data['date'] = pd.to_datetime(data['date'])
data['date'].dt.date.value_counts().sort_index().plot()
plt.title("Distribution of Dates")
plt.xlabel("Date")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# Convert 'title' column to string type
data['title'] = data['title'].astype(str)

title_text = ' '.join(data['title'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(title_text)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title("Word Cloud of Titles")
plt.axis('off')
plt.show()
```
